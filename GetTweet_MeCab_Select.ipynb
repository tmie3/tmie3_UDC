```python
# ライブラリのインポート
import MeCab
import tweepy
```

```python
# 取得した各種キーを格納---------------------------------------------
consumer_key ="キーを入力"
consumer_secret ="キーを入力"
access_token="トークンを入力"
access_token_secret ="トークンを入力"
# Twitterオブジェクトの生成
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
# Mecabのインスタンス
mecab = MeCab.Tagger()
```

```python
#ツイートを取得
Account = "nagoya_kanko" #取得したいユーザーのユーザーIDを代入
tweets = api.user_timeline(Account, count=200, page=1,include_rts=False)
lists = []
for tweet in tweets:
    lists.append(tweet.text)
```

```python
# ツイートをパース
tlist=[]
slist=[]
for x in lists:
    result = mecab.parse(x)
    print(result)
    tlist=result.splitlines()
    for a in tlist:
        if '地名' in a:#抽出したい品詞
            slist.append(a)
    print('='*80) 
```

```python
for b in slist:
     print(b)#抽出したもののみ表示
```
